```python
import numpy as np
import matplotlib.pyplot as plt
SAMPLE_RATE = 100
DURATION = 15

def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(5, duration, sample_rate * duration, endpoint=False)
    y = np.sin((2* np.pi) * freq * x)
    return x,y

t, y = generate_sine_wave(5, SAMPLE_RATE, DURATION)
t1,y1 = generate_sine_wave(2,SAMPLE_RATE,DURATION)

plt.plot(t ,y)
plt.xlabel("t")
plt.ylabel("y")
plt.show()
```

```python
from scipy.fft import fft, fftfreq

N = SAMPLE_RATE * DURATION

yf = fft(y)
f = fftfreq(N, 1/SAMPLE_RATE)

plt.plot(f, np.abs(yf))
plt.xlabel("f in Hz")
plt.ylabel("Amplitude")
plt.show()
```

```python
t_ext = np.arange(0, 2*DURATION, 1/SAMPLE_RATE)
restored_sig = np.zeros(t_ext.size)
for i in range(f.size):
    ampli = np.abs(yf[i]) / N  # amplitude
    phase = np.angle(yf[i])  # phase
    restored_sig += ampli * np.cos(2 * np.pi * f[i] * t_ext + phase)

#plt.plot(t_ext[0:30],restored_sig[0:30])
plt.plot(t[0:30] ,y[0:30])
```
