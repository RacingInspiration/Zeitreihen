```python
# Einlesen des csv als Pandas DataFrame
import pandas as pd
df = pd.read_csv('globallandtemperaturesbymajorcity.csv')
df.head()
```

```python
# Ziel der Verarbeitung: Erstellung eines Graphen mit der Jahresdurchschnittstemperatur in Deutschland
```

```python
# 1. Umwandlung der Zeitspalte in eine echte Datumsspalte
df['dt'].dtype # zuerst liegt gar kein Datumsformat vor!
df['Date'] = pd.to_datetime(df['dt'], format='%Y-%m-%d')
df.head()
```

```python
# 2. Einführung einer Jahresspalte
df['year'] = df['Date'].dt.year
df
```

```python
# 3. Subselektion des Datensatzes
df_Germany = df[df['country'] == 'Germany']
df_Germany
```

```python
# 4. Beseitigen der NAN-Werte bei der Tav 
df_Germany = df_Germany.dropna(subset = 'averagetemperature')
df_Germany
```

```python
# 5. Darstellen der Abhängigkeit Date --> averagetemperature (monatsaufgelöste Temperaturen)
import plotly.express as px

fig = px.line(df_Germany, x="Date", y="averagetemperature", title='Monatstemperaturen')
fig.show()
```

```python
# 6. Zeitauswahl von 1949 bis 2001
df_Germany_Zeitauswahl = df_Germany[(df_Germany['year']>1949) & (df_Germany['year']<=2001)]
df_Germany_Zeitauswahl
```

```python
fig = px.line(df_Germany_Zeitauswahl, x="Date", y="averagetemperature", title='Monatstemperaturen')
fig.show()
```

```python
# 7. Aggregation der Durchschnittstemperaturen auf Jahresbasis
df_Germany_years = df_Germany.groupby('year')['averagetemperature'].aggregate('mean')
df_Germany_years
```

```python
fig = px.line(df_Germany_years, x=df_Germany_years.index, y="averagetemperature", title='Jahrestemperaturen')
fig.show()
```
